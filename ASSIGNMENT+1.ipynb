```python
# Question 5
married = resp[resp["fmarno"] != 0]
neverMarried  = resp[resp["fmarno"] == 0]
```

```python
neverMarried.head()
```

```python
import matplotlib.pyplot as plt

plt.hist(x = [married["totincr"], neverMarried["totincr"]], label = ["Married", "Not Married"])
plt.legend()
plt.show()
```
